```python
#Import Libraries
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector, Operator
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

#define Quantum and classical Registers
q=QuantumRegister(3,'q')
c=ClassicalRegister(3,'c')
qc=QuantumCircuit(q,c)

#Apply H Operator
qc.h(q[0])

# Add barrier with assigned label
qc.barrier(label='After H gate')

# Display the current state after 1st hadmard gate
qc.draw(output='mpl')

# Apply the CNOT operator twice
qc.cx(q[0], q[1])
qc.cx(q[0], q[2])

#Display the unitary operator applied on the qubits until this point
display(Operator(qc).draw('latex'))

# measure the qubit, and store the outcome on the classical bit
qc.measure(q,c)

# execute the circuit
job = AerSimulator().run(qc, shots=10000)
counts = job.result().get_counts(qc)
print(counts)
plot_histogram(counts)

```
