```python
import dimod
from dwave.samplers import SimulatedAnnealingSampler

lamda = 100

Q= {
    (0,0) : 16 - (3*lamda),
    (0,1) : 2 * lamda,
    (0,2) : 2 * lamda,
    (1,1) : 22 - (3*lamda),
    (1,2) : 2 * lamda,
    (2,2) : 14 - (3*lamda),
}

bqm = dimod.BinaryQuadraticModel.from_qubo(Q)

sampler = SimulatedAnnealingSampler()
sampletest = sampler.sample(bqm, num_reads=10)

for sample, energy, num_occurrences in sampletest.data(['sample', 'energy', 'num_occurrences']):
    print(f"Sample: {sample}, Energy: {energy}, Occurrences: {num_occurrences}")
```

```python
import dimod
from dwave.samplers import SimulatedAnnealingSampler
from matplotlib import pyplot as plt

lamda =10

Q= {
    (1,1) : 1,
    (2,2) : 2,
    (3,3) : 3,
    (1,2) : 3 * lamda,
    (1,3) : -1 * lamda,
}

bqm = dimod.BinaryQuadraticModel.from_qubo(Q)

sampler = SimulatedAnnealingSampler()
sampletest = sampler.sample(bqm, num_reads=10)

for sample, energy in sampletest.data(['sample', 'energy']):
    print(f"Sample: {sample}, Energy: {energy})")
```
